```python
import datetime
since = datetime.datetime(2022, 3, 4)
until = datetime.datetime(2022, 3, 18)#(2022, 3, 15)
print(since)
print(until)
```

```python
#configuration
config = twint.Config()
config.Search = "#NFTs NFTCommunity"
config.Lang = "en"
config.Limit = 2000
config.Since = str(since)
config.Until = str(until)
config.Store_json = True
config.Output = "g.json"
#running search
twint.run.Search(config)
```

```python
data = pd.read_json('g.json', lines=True)
```
